```python
import numpy as np

images = []

for i in range(9):
    x = np.load("images/car_"+ str(i) + ".npy")
    images.append(x)
images = np.array(images)
print(images.shape)
```

```python
np.sum(images)
```

```python
suma = []

for img in images:
    suma.append(np.sum(img))
print(suma)
```

```python
max = 0

for i, s in enumerate(suma):
    if(s > max):
        max = s
        imax = i
print(imax)
```

```python
from skimage import io

mean_img = np.mean(images, axis = 0)
io.imshow(mean_img.astype(np.uint8)) # petru a putea fi afisata
                                   # imaginea trebuie sa aiba
                                   # tipul unsigned int
io.show()
```

```python
dev = np.std(images)
print(dev)
```

```python
normalizare = (images - mean_img) / dev
print(normalizare)
```
